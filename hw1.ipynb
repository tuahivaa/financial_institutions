```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = "current.csv"
df = pd.read_csv(dataset)
```

UTILITY FUNCTIONS

```python
def table(df, start, end):
    group_data = df.iloc[start-1:end]
    stats = group_data.describe(percentiles=[.25, .5, .75])
    stats = stats.transpose()
    selected_columns = ['mean', '25%', '50%', '75%', 'std', 'min', 'max']
    table = stats[selected_columns]
    return table

def heatmap(table):
    correlation = table
    plt.figure(figsize=(10,8))
    sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title('Heatmap')
    plt.show()
```

GROUP 1 - Output and Income

```python
table_1 = table(df, start=1, end=20)
print(table_1)
heatmap(table_1)
```

GROUP 2 - Labor Market

```python
table_2 = table(df, start=21, end=53)
print(table_2)
heatmap(table_2)
```
